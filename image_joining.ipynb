```python


# Load the images
img1 = Image.open("results_t_3/vis_DC_MNIST_MLP_1ipc_single_iter0.png")
img2 = Image.open("results_t_3/vis_DC_MNIST_MLP_1ipc_single_iter5.png")
img3 = Image.open("results_t_3/vis_DC_MNIST_MLP_1ipc_single_iter10.png")
img4 = Image.open("results_t_3/vis_DC_MNIST_MLP_1ipc_single_iter15.png")
img5 = Image.open("results_t_3/vis_DC_MNIST_MLP_1ipc_single_iter20.png")
img6 = Image.open("results_t_3/vis_DC_MNIST_MLP_1ipc_single_iter25.png")

# Get dimensions for the final image
width, height = img1.size
total_width = width * 6  # Since we are stacking 6 images horizontally
combined_image = Image.new("RGB", (total_width, height))

# Paste images next to each other
combined_image.paste(img1, (0, 0))
combined_image.paste(img2, (width, 0))
combined_image.paste(img3, (2 * width, 0))
combined_image.paste(img4, (3 * width, 0))
combined_image.paste(img5, (4 * width, 0))
combined_image.paste(img6, (5 * width, 0))

# Save the final combined image
combined_image.save("MLP_Hess_combined_image_with_iter25.png")
combined_image.show()  # Opens the image in the
```
